```python
# Define a function to calculate rankings based on position
def calculate_position_rankings(df):  
    # Initialize the 'Points' column
    df["Total Points"] = 0
    
    # Loop through rows and calculate points based on position group
    for index, row in df.iterrows():
        position_group = row["pos"]
        
        # Select appropriate weights
        if position_group == "FW" or "FW,MF":
            weights = column_weights_fw
        elif position_group == "MF":
            weights = column_weights_mf
        else:
            weights = column_weights_df
        
        for c, weight in weights.items():
            df['Total Points'] += weight * (num_players - df[c].rank(ascending = False) + 1)
    
    return df

# Apply the function to calculate rankings
ranked_table = calculate_position_rankings(
    bd_stats_cleaned
)

# Sort by position and rank for display
ranked_table = ranked_table.dropna()
ranked_table = ranked_table.sort_values(by="Total Points", ascending = False)
ranked_table["Rank"] = ranked_table["Total Points"].rank(ascending=False, method="min").astype(int)

# Sort the DataFrame by Rank
ranked_table = ranked_table.sort_values(by="Rank")

# Set the Rank column as the new index
ranked_table.set_index("Rank", inplace=True)
ranked_table
```
